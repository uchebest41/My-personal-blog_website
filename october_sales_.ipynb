```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload the Excel file to Google Colab
uploaded = files.upload()

# Assuming the uploaded file is named 'merged_file (4).xlsx'
uploaded_filename = next(iter(uploaded))

# Read the uploaded Excel file
df = pd.read_excel(uploaded_filename)

# Create a pivot table for 'AMOUNT', 'Cost', '% M/S', 'Profit'
pivot_table = df.pivot_table(
    values=['AMOUNT', 'Cost', '% M/S', 'Profit'],
    index='File_Name',
    aggfunc='sum'
)

# Display the pivot table
display(pivot_table)

# Plot a line chart for 'Profit' for each 'File_Name'
plt.figure()
for file_name, group in df.groupby('File_Name'):
    plt.plot(group['Profit'], label=file_name)

plt.title('Profit for Each File_Name')
plt.xlabel('Index')
plt.ylabel('Profit')
plt.legend()

# Save both the pivot table and the line chart to the same Excel file
with pd.ExcelWriter(uploaded_filename, engine='openpyxl', mode='a') as writer:
    # Save the pivot table to a new sheet
    pivot_table.to_excel(writer, sheet_name='Pivot_Table')

    # Create a new sheet for the line chart
    line_chart_sheet = writer.book.create_sheet('Line_Chart')

    # Save the line chart as an image in the new sheet
    img = openpyxl.drawing.image.Image('/content/line_chart.png')
    img.anchor = 'A1'
    line_chart_sheet.add_image(img)

# Download the modified Excel file
files.download(uploaded_filename)
```
